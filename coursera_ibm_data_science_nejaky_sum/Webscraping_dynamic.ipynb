```python
from selenium import webdriver
from bs4 import BeautifulSoup
import time

def clik_to_property(selektor, realita):
    typy_realit = driver.find_elements_by_css_selector(str(selektor))   
    for reality in typy_realit:
        if reality.text == str(realita):
            return reality.click()
            
def filter_contract (selektor,typ_kontrakt):
    contracts = driver.find_elements_by_css_selector(str(selektor))
    for contract in contracts:
        if contract.text == str(typ_kontrakt):
            return contract.click()
        
def filter_land (selektor,vyber_zeme):
    zeme_list = driver.find_elements_by_css_selector (str(selektor))
    for zeme in zeme_list:
        if zeme.text == str(vyber_zeme):
            return zeme.click()
        
# li.item ng-binding ng-scope active move
web_driver = r'c:\Users\jan.petr\OneDrive - Ipsos\Provozni\chromedriver\chromedriver.exe'
stranka = 'https://www.sreality.cz'


driver = webdriver.Chrome(web_driver)
driver.get(stranka)
    
            
time.sleep(3) # Let the user actually see something!

###klik na domy
clik_to_property('a.dir-hp-signpost__item__link', "Domy")     
###klik na kontrakt
filter_contract ('a.switcher',"Prodej")
time.sleep(3)
###vyber zeme
filter_land ("div.locality-country-id option", "Česká republika")
time.sleep(3) 


tlacitko_hledej = driver.find_element_by_css_selector('button.btn-XL')
tlacitko_hledej.click()
time.sleep(5)
soup = BeautifulSoup(driver.page_source)

#click_domy = typy_realit[1].find_element_by_link_text('Domy')
###search_box = driver.find_element_by_name('input')
###search_box.send_keys('ChromeDriver')
###search_box.submit()
time.sleep(5) # Let the user actually see something!
driver.quit()
```
